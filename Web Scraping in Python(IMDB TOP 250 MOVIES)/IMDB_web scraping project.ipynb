```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd

```

<h2> Requestes page sources</h2>
<p> Go to the targeted website , Press F12 ,select neworks and reload the page. Select the top file and grab the user agent from http request section</p>

```python
url='https://www.imdb.com/chart/top/'


headers = {
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8',
    'Accept-Encoding': 'gzip, deflate, br, zstd',
    'Accept-Language': 'en-US,en;q=0.6',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'
}

page = requests.get(url,headers=headers)
             
```

```python
soup = BeautifulSoup(page.content, 'lxml')
```

```python
movie_tags = soup.find_all("h3", class_="ipc-title__text")
movie_tags

#scrapped_movies = soup.select('li.ipc-metadata-list-summary-item.sc-10233bc-0.iherUv.cli-parent div div div a')
#scrapped_movies
# Display the scrapped movies
#for movie in scrapped_movies:
#    print(movie.text, movie['href'])
#//li[@class='ipc-metadata-list-summary-item sc-10233bc-0 iherUv cli-parent']//div//div//div//a

```

```python
movie_names = [tag.text.split('. ', 1)[-1] for tag in movie_tags if ". " in tag.text]

movie_names_list = []

for name in movie_names:
    movie_names_list.append(name)
print(movie_names_list)
```

```python
df = pd.DataFrame(movie_names_list, columns=["Movie Names"])
# Print the DataFrame
df.head()
```

```python
movie_rating=soup.find_all('span', class_='ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating')
movie_rating
```

```python
movie_ratings = [float(tag['aria-label'].split(': ')[-1]) for tag in movie_rating]
movie_ratings_list=[]

for rate in movie_ratings:
    movie_ratings_list.append(rate)
print(movie_ratings_list)
```

```python
df['Ratings'] = movie_ratings
df
```
